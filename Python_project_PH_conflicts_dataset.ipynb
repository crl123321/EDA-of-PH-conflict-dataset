```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pr = pd.read_csv('conflict_data_phl.csv')
pr.head()
```

```python
pr.info()
pr.isnull().sum()
```

```python
pr.describe(exclude='number')
```

```python
pr.drop(pr.index[0], inplace=True)
pr.head()
```

```python
years = pr.groupby('year').count()['id'].reset_index() 
years
```

```python
fig = plt.figure(figsize = (25,10))
sns.lineplot(x = "year", y = "id", data = years, marker = "o")
plt.title("Number of Cinflicts per Year")
plt.xlabel("")
plt.ylabel("")
plt.style.use("fivethirtyeight")
plt.show()
```

```python
most = years.sort_values(by = "id", ascending = False)
most
```

```python
year = most["year"]
id = most["id"]

fig = plt.figure(figsize = (25,10))
plt.barh(year[::-1], id[::-1])
plt.title("Number of Conflicts from 1989 to 2021", loc = "center")

plt.show()
```

```python
province = pr.groupby('adm_1').count()['id'].reset_index().sort_values(by = 'id', ascending=False)
province
```

```python
adm = province['adm_1']
ids = province['id']

fig = plt.figure(figsize = (25,25))
plt.barh(adm[::-1], ids[::-1])
plt.title("Number of Conflicts per Province", loc = "center")

plt.show()
```

```python
mun = pr.groupby(['year', 'adm_2']).count()['id'].reset_index().sort_values(by = 'id', ascending=False).head(10)
mun
```

```python
pivot1 = mun.pivot('adm_2', 'year', 'id').head(10)
pivot1
```

```python
mun2 = pr.groupby('adm_2').count()['id'].reset_index().sort_values(by = 'id', ascending=False).head(10)
mun2
```

```python
adm2 = mun2['adm_2']
idss = mun2['id']

fig = plt.figure(figsize = (25,25))
plt.barh(adm2[::-1], idss[::-1])
plt.title("Number of Conflicts per Municipality", loc = "center")

plt.show()
```
